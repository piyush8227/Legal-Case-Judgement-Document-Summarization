```python
data_1 = data.copy()
```

```python
data_1.info()
```

```python
data_1.dropna(subset=['JUDGMENT'], inplace=True)
```

```python
data_1.head()
```

```python
data_1.reset_index(inplace=True)
```

```python
data_1.drop(columns='index', inplace=True)
```

```python
indexes = []
for i in range(len(data_1['JUDGMENT'])):

  if 4091 > len(data_1['JUDGMENT'][i]) > 0:
    indexes.append(i)
```

```python
!pip install transformers
```

```python
from transformers import pipeline

summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
```

```python
summarized = []
for index in indexes:
    print(index)
    ARTICLE = data_1['JUDGMENT'][index]
    try:
        summary = summarizer(ARTICLE, max_length=4096, min_length=30, do_sample=False)
        summarized.append(summary[0]['summary_text'])
    except IndexError:
        print(f"Judgment at index {index} surpasses the model's input size. Skipping...")
        continue
```

```python
counts = [374, 434, 619, 853, 854, 880, 1046, 1428, 2749, 3147, 3444, 3955, 4330, 4791, 5085,
         7212, 8287, 8441, 8469, 9534, 9627,9666, 9695, 10076, 10408, 10924, 11031, 11590,
         11787, 11982, 11994, 12507, 12654, 12760, 12902, 12977, 12987, 13120, 13143, 13716,
         13847, 14009, 15355, 15393, 15603, 15801, 16444, 16620, 17196, 17314, 17732, 17795,
         17951, 18791, 19262, 19317, 19520, 19589, 19973, 20135, 20298, 20812, 21384,
         21677, 21844, 22454, 22588, 22931, 23199, 23693, 24465, 24748]
```

```python
len(counts)
```

```python
judgements = []
for index in indexes:
    if index not in counts:
        ARTICLE = data_1['JUDGMENT'][index]
        judgements.append(ARTICLE)
```

```python
len(judgements)
```

```python
data_2 = pd.DataFrame(judgements, columns = ['JUDGMENT'])
```

```python
data_2['Summarized_text'] = summarized
```
