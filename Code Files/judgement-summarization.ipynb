```python
data_2 = data.iloc[:680,:]
```

```python
from transformers import pipeline

pipe = pipeline(model="facebook/bart-large-mnli")
```

```python
labels = []
scores = []
for i in range(data_2.shape[0]):
    print(i)
    result = pipe(data_2['JUDGMENT'][i],
                  candidate_labels=["criminal", "divorce", "civil", "transfer", 
                  "barge hire charges", "appointment", "special leave petitions"],)
    labels.append(result['labels'][0])
    scores.append(round(result['scores'][0]*100, 2))
```

```python
new_data_1 = pd.DataFrame({'Label':labels, 'Score': scores})
```

```python
data_3 = data.iloc[680:1360,:]
```

```python
labels_2 = []
scores_2 = []
for i in range(680, 1360):
    print(i)
    result = pipe(data_3['JUDGMENT'][i],
                  candidate_labels=["criminal", "divorce", "civil", "transfer", 
                  "barge hire charges", "appointment", "special leave petitions"],)
    labels_2.append(result['labels'][0])
    scores_2.append(round(result['scores'][0]*100, 2))
```

```python
new_data_2 = pd.DataFrame({'Label':labels_2, 'Score': scores_2})
```

```python
new_data_1
```

```python
data_4 = pd.concat([new_data_1, new_data_2])
```

```python
data_4.reset_index(inplace=True)
```

```python
data1 = data.iloc[:1360,:]
```

```python
data1['Label'] = data_4['Label']
```

```python
data1['Scores'] = data_4['Score']
```

```python
data1.to_csv('New_Labeled_Data.csv',index=False)
```
